```python
d1 = {}
for item in df_modified['Model Name'].unique():
  d1[item] = item.split()[0].lower()
```

**importing 101 csv file**

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/Nisarg221B/Electric_Vehicle_Market_Segmentation/master/collected%20datasets/101.csv', sep=',', encoding='cp1252').iloc[:-1, :]
```

```python
df1.head()
```

```python
d = {}
for item in df1['OEM’s Registered']:
  d[item] = item[:-1]
```

```python
df1['OEM’s Registered'] = df1['OEM’s Registered'].map(d)
```

```python
df1
```

```python
d2 ={}
for item in df1['OEM’s Registered']:
  list_item = item.split(',')
  for x in list_item:
    d2[x.split()[0].lower()] = df1.loc[df1['OEM’s Registered'] == item][['Category of e-Vehicle']].values[0][0]
```

```python
d2
```

```python
df_modified['Model Name'].unique()
```

```python
df_modified['Category of e-Vehicle'] = df_modified['mm'].map(d2)
```

```python
df_modified
```

```python
df_modified.drop(labels='mm', axis=1, inplace=True)
```
